```python
import requests
from bs4 import BeautifulSoup
```

# Loding Website to Python

```python
response = requests.get('https://finance.yahoo.com/most-active/')
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    # print(soup.prettify())
else:
    print('No Response.')
```

# Data Extraction

```python
table = soup.find('table', class_ = 'W(100%)')
# print(table)
```

```python
thead = table.find('thead')
# print(thead)
th = thead.find_all('th')
# print(th)
```

Extracting Headers

```python
headers = []
for h in th:
    header = str(h.text).strip()
    headers.append(header)

print(headers)
    
```

Extracting Column data

```python
tbody = table.find('tbody')
allrows = tbody.find_all('tr')
print(allrows)
```

```python
# print(len(allrows))
alltd = []
index = 0
while index >= 0:
    tr = allrows[index]
    # print(tr)
    list = []
    for td in tr:
        data = td.text
        # print(data)
        list.append(data)
    # print(list)
    alltd.append(list)
    # print(index)
    index += 1
    if index == len(allrows):
        break
# print(alltd)
    
```

Cleaning

```python
# print(headers)

headers.pop()
# print(headers)

# print(alltd[0])

rowdata = []
for i in alltd:
    i.pop()
    # print(i)
    rowdata.append(i)  
# print(rowdata)
```

Validation

```python
print(len(headers))
print(len(alltd[0]))
```

```python
# headers
data_rows = rowdata
# print(headers)
# print(data_rows)
```

```python
import pandas as pd
df = pd.DataFrame(data_rows, columns=headers)
print(df)
```

```python
df.to_csv('Stock Market Data.csv', index= False)
```
